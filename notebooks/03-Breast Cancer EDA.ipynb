## Univariate Plots

```python
sns.countplot(x='diagnosis', data=df); 
```

```python
# histograms 
df.hist(figsize=(20,10), color="#5F9EA0")
plt.show() 
```

```python
# density plots 
df.plot(kind='density', subplots=True, layout=(6,6), sharex=False, figsize=(20,10))
plt.show() 
```

```python
# boxplots  
df.plot(kind='box', subplots=True, layout=(6,6), sharex=False, figsize=(20,10))
plt.show() 
```

```python
# correlation plots 
fig,ax = plt.subplots(figsize=(18,18))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show() 
```

```python
# pair plots 
sns.pairplot(df)
plt.show() 
```
